```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("output.outfmt6", sep="\t", header=None)
```

```python
df.head(10)
```

```python
df[2] > 50
```

```python
id_gr_50 = df[df[2] > 50]
```

```python
id_gr_50.head()
```

```python
id_gr_50.to_csv("filtered_output.outfmt6", sep="\t", index=False)
```

```python
plt.figure(figsize=(6, 4))
id_gr_50[2].hist(bins=10)
plt.title("Distribution of Identity Values")
plt.xlabel("Identity (%)")
plt.ylabel("Frequency")
plt.savefig("id_dist_hist.png")
plt.show()
```

```python
id_gr_50[5].value_counts(ascending=False)
```

```python
plt.figure(figsize=(12, 4))
taxonomy_counts = id_gr_50[5].value_counts()
taxonomy_counts.plot(kind="bar")
plt.title("Taxonomy Diversity")
plt.xlabel("Scientific Name")
plt.ylabel("Count")
plt.savefig("taxonomy_barplot.png")
plt.show()
```
